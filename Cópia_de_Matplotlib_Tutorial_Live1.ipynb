```
import numpy as np
import matplotlib.pyplot as plt
```

```
x=np.linspace(0,20,1000)
y=np.sin(x)+0.2*x
plt.plot(x,y)
```

```
x=np.linspace(0,20,1000)
y=np.sin(x)+0.2*x
plt.plot(x,y)
plt.show()
```

```
x=np.linspace(0,20,1000)
y=np.sin(x)+0.2*x
plt.plot(x,y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("MyPlot")
```

```
#scatterplot
x=np.random.randn(100,2)
plt.scatter(x[:,0],x[:,1])
```

```
x=np.random.randn(200,2)
x[:50]+=3
y=np.zeros(200)
y[:50]=1
plt.scatter(x[:0],x[:,1], c=y)
```

```
#Histograma
x=np.random.randn(10000)
plt.hist(x);
plt.hist(x,bins=50);
```
